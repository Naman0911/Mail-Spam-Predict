# Independent and Dependent Split

```python
X = df.drop('spam',axis = 1)
Y = df['spam']
X
```

```python
Y
```

# Train test split

```python
from sklearn.model_selection import train_test_split

X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.25 , random_state=42 , stratify=Y)
```

# Model Training

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score , classification_report , confusion_matrix , recall_score , precision_score , roc_curve , roc_auc_score , f1_score

models = {
    "Adaboost" : AdaBoostClassifier(),
    "Random Forest" : RandomForestClassifier(),
    "Decision Trees" : DecisionTreeClassifier(),
    "Logistic Regression" : LogisticRegression(),
    "SVM" : SVC(),
    "KNN" : KNeighborsClassifier()
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    
    model.fit(X_train , Y_train)
    
    Y_pred_train = model.predict(X_train)
    Y_pred_test = model.predict(X_test)
    
    train_accuracy = accuracy_score(Y_train , Y_pred_train)
    train_recall = recall_score(Y_train , Y_pred_train)
    train_precision = precision_score(Y_train , Y_pred_train)
    train_f1 = f1_score(Y_train , Y_pred_train)
    train_roc_auc_score = roc_auc_score(Y_train , Y_pred_train)
    train_report = classification_report(Y_train , Y_pred_train)
    train_matrix = confusion_matrix(Y_train , Y_pred_train)
    
    test_accuracy = accuracy_score(Y_test , Y_pred_test)
    test_recall = recall_score(Y_test , Y_pred_test)
    test_precision = precision_score(Y_test , Y_pred_test)
    test_f1 = f1_score(Y_test , Y_pred_test)
    test_roc_auc_score = roc_auc_score(Y_test , Y_pred_test)
    test_report = classification_report(Y_test , Y_pred_test)
    test_matrix = confusion_matrix(Y_test, Y_pred_test)
    
    
    print(f'#-------------------- {list(models.keys())[i]} --------------------#')
    
    print("Training Accuracy =",train_accuracy)
    print("Training Recall =",train_recall)
    print("Training Precision =",train_precision)
    print("Training f1 =",train_f1)
    print("Training ROC_AUC_Score =",train_roc_auc_score)
    print(train_report)
    print(train_matrix)
    
    print("-"*50)
    
    print("Test Accuracy =",test_accuracy)
    print("Test Recall =",test_recall)
    print("Test Precision =",test_precision)
    print("Test f1 =",test_f1)
    print("Test ROC_AUC_Score =",test_roc_auc_score)
    print(test_report)
    print(test_matrix)
    
    print('='*50)
    
    print('\n')
```

# Observation
- From Above Model Training we got the best models which are the Random forest and ADAboost

# Hyperparameter Tuning For the Adaboost

```python
ada_params = {
    'n_estimators' : [100,200,500,1000],
    'algorithm' : ['SAMME','SAMME.R']
}

RandCV = [("AdaBoost",AdaBoostClassifier(),ada_params)]


from sklearn.model_selection import RandomizedSearchCV
model_name = {}
for name,model,param in RandCV:
    
    rand = RandomizedSearchCV(estimator=model , param_distributions=param , cv = 3 , n_jobs = -1 , n_iter=100)
    rand.fit(X_train , Y_train)
    model_name[name] = rand.best_params_
    
for model_ in model_name:
    print(f"# ----------------------- Best Params for {model_} --------------------------- #")
    print(model_name[model_])
```

```python
from sklearn.ensemble import AdaBoostClassifier

from sklearn.metrics import accuracy_score , classification_report , confusion_matrix , recall_score , precision_score , roc_curve , roc_auc_score , f1_score

models = {
    "Adaboost" : AdaBoostClassifier(n_estimators = 1000 , algorithm= 'SAMME' )
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    
    model.fit(X_train , Y_train)
    
    Y_pred_train = model.predict(X_train)
    Y_pred_test = model.predict(X_test)
    
    train_accuracy = accuracy_score(Y_train , Y_pred_train)
    train_recall = recall_score(Y_train , Y_pred_train)
    train_precision = precision_score(Y_train , Y_pred_train)
    train_f1 = f1_score(Y_train , Y_pred_train)
    train_roc_auc_score = roc_auc_score(Y_train , Y_pred_train)
    train_report = classification_report(Y_train , Y_pred_train)
    train_matrix = confusion_matrix(Y_train , Y_pred_train)
    
    test_accuracy = accuracy_score(Y_test , Y_pred_test)
    test_recall = recall_score(Y_test , Y_pred_test)
    test_precision = precision_score(Y_test , Y_pred_test)
    test_f1 = f1_score(Y_test , Y_pred_test)
    test_roc_auc_score = roc_auc_score(Y_test , Y_pred_test)
    test_report = classification_report(Y_test , Y_pred_test)
    test_matrix = confusion_matrix(Y_test, Y_pred_test)
    
    
    print(f'#-------------------- {list(models.keys())[i]} --------------------#')
    
    print("Training Accuracy =",train_accuracy)
    print("Training Recall =",train_recall)
    print("Training Precision =",train_precision)
    print("Training f1 =",train_f1)
    print("Training ROC_AUC_Score =",train_roc_auc_score)
    print(train_report)
    print(train_matrix)
    
    print("-"*50)
    
    print("Test Accuracy =",test_accuracy)
    print("Test Recall =",test_recall)
    print("Test Precision =",test_precision)
    print("Test f1 =",test_f1)
    print("Test ROC_AUC_Score =",test_roc_auc_score)
    print(test_report)
    print(test_matrix)
    
    print('='*50)
    
    print('\n')
```

# ROC_CURVE

```python
plt.figure()

# Add the models to the list that we want to view on the ROC plot 

auc_models = [{
    'label':'AdaBoost Classifier',
    'model':AdaBoostClassifier(n_estimators=1000 , algorithm = 'SAMME'),
    'auc' : 0.9385973820807944
    },]

# Create loop through all models

for algo in auc_models:
    model = algo['model']              # Select the model
    model.fit(X_train , Y_train)       # Train the dataset
    
    # compute FPR TPR and threshold
    fpr , tpr , thresholds = roc_curve(Y_test , model.predict_proba(X_test)[:,1])
    
    # Calculating Area under Curve
    plt.plot(fpr,tpr,label = '%s ROC (area = %0.2f)' % (algo['label'] , algo['auc']))
    
    plt.plot([0,1],[0,1],'r--')
    plt.xlim([0.0,1.0])
    plt.ylim([0.0,1.05])
    plt.xlabel("1 - Specificity(False Positive Rate)")
    plt.ylabel("Sensitivity(True Positive Rate)")
    plt.title("Receiver Operating Characteristic")
    plt.legend(loc="lower right")
    plt.savefig("auc.png")
    plt.show()
```
